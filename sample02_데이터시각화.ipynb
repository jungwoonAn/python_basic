```python
import matplotlib.pyplot as plt

plt.title('Graph Title')
plt.plot([1,2,3,4], [1,4,9,16], label='power')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
x_val = [100,200,300]
tmp_val = ['A-Store','B-Store','C-Store']
# color : r, g, b, c, m, y ...
plt.bar(x, x_val, color='r')
plt.xticks(x, tmp_val)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(3)
y_val = [100,200,300]
plt.barh(y, y_val, color='b')
plt.show()
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np

# np.random.seed(0)  # 고정값
np.random.seed(datetime.now().microsecond)
n = 30
x = np.random.rand(n)
y = np.random.rand(n)
s_marker = (20*np.random.rand(n))**2
col = np.random.rand(n)
plt.scatter(x, y, s=s_marker, c=col)
plt.show()
```
